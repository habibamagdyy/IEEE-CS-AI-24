The cars in this dataset are categorized into one of five different vehicle classes based on size. Starting from the smallest, they are: `{Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, and Large Cars}`.

**Hint**: Make sure that the order of vehicle classes makes sense in your plot!

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'fuel-econ.csv'
data = pd.read_csv(file_path)

```

```python
# necessary columns
vehicle_classes = data['VClass']
engine_sizes = data['displ']

order = ['Minicompact Cars', 'Subcompact Cars', 'Compact Cars', 'Midsize Cars', 'Large Cars']
```

```python
# plot
plt.figure(figsize=(12, 8))
sns.boxplot(x=vehicle_classes, y=engine_sizes, order=order)
plt.title('Engine Size by Vehicle Class')
plt.xlabel('Vehicle Class')
plt.ylabel('Engine Size (L)')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
