# Update coulmn values in Marital Status to a charcheter crossponding it's value
for example Single ->S ,Married->M and so on

```python
marital_mapping = {'Single': 'S', 'Married': 'M', 'Divorced': 'D'}
alumni['Marital Status'] = alumni['Marital Status'].map(marital_mapping)
```

# Drop the Fee coulmn

```python
alumni.drop(columns=['Fee'], inplace=True)
```
