### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv).

### Step 3. Assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, delimiter='\t')
print(chipo.head())
```

### Step 4. See the first 10 entries

```python
print(chipo.head(10))
```

### Step 5. What is the number of observations in the dataset?

```python
#number of observations
num_observations = chipo.shape[0]
print("Number of observations in the dataset:", num_observations)
```

```python
#show info about the data
print(chipo.info())
```

### Step 6. What is the number of columns in the dataset?

```python
num_columns = chipo.shape[1]
print("Number of columns in the dataset:", num_columns)
```

### Step 7. Print the name of all the columns.

```python
print(chipo.columns)
```

### Step 8. How is the dataset indexed?

```python
print(chipo.index)
```

### Step 9. Which was the most-ordered item?

```python
MOI = chipo.groupby('item_name')['quantity'].sum().idxmax()
print("Most-ordered item:", MOI)
```

### Step 10. For the most-ordered item, how many items were ordered?

```python
quantity_ordered = chipo.groupby('item_name')['quantity'].sum().max()
print("Number of items ordered for the most-ordered item:", quantity_ordered)
```

### Step 11. What was the most ordered item in the choice_description column?

```python
most_ordered_choice = chipo.groupby('choice_description')['quantity'].sum().idxmax()
print("Most ordered item in the choice_description column:", most_ordered_choice)
```

### Step 12. How many items were orderd in total?

```python
total_items_ordered = chipo['quantity'].sum()
print("Total number of items ordered:", total_items_ordered)
```

### Step 13. Turn the item price into a float

#### Step 13.a. Check the item price type

```python
print("Type of item price before conversion:", chipo['item_price'].dtype)
```

#### Step 13.b. Create a lambda function and change the type of item price

```python
chipo['item_price'] = chipo['item_price'].apply(lambda x: float(x[1:]))
```

#### Step 13.c. Check the item price type

```python
print("Type of item price after conversion:", chipo['item_price'].dtype)
```

### Step 14. How much was the revenue for the period in the dataset?

```python
revenue = (chipo['quantity'] * chipo['item_price']).sum()
print("Revenue for the period:", revenue)
```

### Step 15. How many orders were made in the period?

```python
num_orders = chipo['order_id'].nunique()
print("Number of orders made in the period:", num_orders)
```

### Step 16. What is the average revenue amount per order?

```python
average_revenue_per_order = revenue / num_orders

```

```python
print("Average revenue amount per order:", average_revenue_per_order)
```
