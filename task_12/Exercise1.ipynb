### Step 1. Import the necessary libraries

```python
import pandas as pd
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv).

### Step 3. Assign it to a variable called drinks.

```python
dataset = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/drinks.csv"
drinks = pd.read_csv(dataset)
```

### Step 4. Which continent drinks more beer on average?

```python
beer_avg = drinks.groupby('continent')['beer_servings'].mean()
continent_most_beer = beer_avg.idxmax()
print(f"The continent that drinks the most beer on average is {continent_most_beer}")
```

### Step 5. For each continent print the statistics for wine consumption.

```python
wine = drinks.groupby('continent')['wine_servings'].describe()
print("\nStatistics for wine consumption by continent:")
print(wine)
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
print(drinks.dtypes)
numeric_columns = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
drinks[numeric_columns] = drinks[numeric_columns].apply(pd.to_numeric, errors='coerce')
try:
    alcohol_means = drinks.groupby('continent').mean()
    print("\nMean alcohol consumption per continent for every column:")
    print(alcohol_means)
except Exception as e:
    print(f"Error occurred: {e}")
```

### Step 7. Print the median alcohol consumption per continent for every column

```python
print(drinks.dtypes)
numeric_columns = ['beer_servings', 'spirit_servings', 'wine_servings', 'total_litres_of_pure_alcohol']
drinks[numeric_columns] = drinks[numeric_columns].apply(pd.to_numeric, errors='coerce')
try:
    alcohol_medians = drinks.groupby('continent').median()
    print("\nMean alcohol consumption per continent for every column:")
    print(alcohol_medians)
except Exception as e:
    print(f"Error occurred: {e}")
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks['spirit_servings'] = pd.to_numeric(drinks['spirit_servings'], errors='coerce')
drinks.dropna(subset=['spirit_servings'], inplace=True)

spirit_stats = drinks.groupby('continent')['spirit_servings'].agg(['mean', 'min', 'max'])
spirit_stats_df = pd.DataFrame(spirit_stats)

print("Mean, min, and max values for spirit consumption:")
print(spirit_stats_df)
```
